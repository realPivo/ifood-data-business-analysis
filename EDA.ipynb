```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('ml_project1_data.csv', index_col='ID')
```

```python
df
```

```python
df.info()
```

```python
df.describe()
```

```python
# Standardization of column names.
df.rename(columns={'MntMeatProducts': 'MntMeat', 'MntFishProducts': 'MntFish', 'MntSweetProducts': 'MntSweet', 'MntGoldProds': 'MntGold'}, inplace=True)
```

```python
# Dropping unknown columns.
df.drop(columns=['Z_CostContact', 'Z_Revenue', 'Response'], inplace=True)
```

```python
# Null handling:
NumInitialRecords = len(df.index)
NumDroppedRecords = len(df.index) - len(df.dropna().index)
print('Percentage of records deleted = {}%'.format(NumDroppedRecords*100/NumInitialRecords))
```

```python
df.dropna(inplace=True)
```

```python
# Change data types for columns.
df['Education'] = df['Education'].astype('category')
df['Marital_Status'] = df['Marital_Status'].astype('category')
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
```

```python
df['AcceptedCmp1'] = df['AcceptedCmp1'].astype(bool)
df['AcceptedCmp2'] = df['AcceptedCmp2'].astype(bool)
df['AcceptedCmp3'] = df['AcceptedCmp3'].astype(bool)
df['AcceptedCmp4'] = df['AcceptedCmp4'].astype(bool)
df['AcceptedCmp5'] = df['AcceptedCmp5'].astype(bool)
```

```python
df.to_csv('prepared_data.csv', index=True)
```

```python
# Correlation analysis
correlation_matrix = df.corr()
```

```python
products_corr = correlation_matrix[['MntWines', 'MntFruits', 'MntMeat', 'MntFish', 'MntSweet', 'MntGold']].reset_index()
products_corr = products_corr.loc[products_corr['index'].isin(('Year_Birth', 'Income', 'Kidhome', 'Teenhome', 'Recency'))].set_index('index')
plt.figure(figsize=(10,5))
sns.heatmap(products_corr, annot=True)
```

We can see a high dependence of the amount spend on products on the income

```python
purchases_corr = correlation_matrix[['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth']].reset_index()
purchases_corr = purchases_corr.loc[purchases_corr['index'].isin(('Year_Birth', 'Income', 'Kidhome', 'Teenhome', 'Recency'))].set_index('index')
plt.figure(figsize=(10,5))
sns.heatmap(purchases_corr, annot=True)
```

There are a high dependencies on:
- number of purchases made with discount / teen in home
- number of purchases on web site / income
- number of purchases made using catalogue / income
- number of purchases made directly in store / income
- number of visits to company's web site / kid in home

```python
campaigns_corr = correlation_matrix[['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']].reset_index()
campaigns_corr = campaigns_corr.loc[campaigns_corr['index'].isin(('Year_Birth', 'Income', 'Kidhome', 'Teenhome', 'Recency'))].set_index('index')
plt.figure(figsize=(10,5))
sns.heatmap(campaigns_corr, annot=True)
```

There is a relatively high dependency on Campaigns 1 & 5 / income

```python
# Success of marketing campaigns
for cmp in ['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5']:
    customers = len(df.index)
    accepted_customers = len(df.loc[df[cmp]] == True)
    pct = accepted_customers * 100 / customers
    print('{} - {}%'.format(cmp, pct))
    
```
